```python
import cv2
import numpy as np

# Global variables
current_player = 'X'  # Player X starts the game
board = [['' for _ in range(3)] for _ in range(3)]  # Empty board
winning_combinations = [
    [(0, 0), (0, 1), (0, 2)],  # Row 1
    [(1, 0), (1, 1), (1, 2)],  # Row 2
    [(2, 0), (2, 1), (2, 2)],  # Row 3
    [(0, 0), (1, 0), (2, 0)],  # Column 1
    [(0, 1), (1, 1), (2, 1)],  # Column 2
    [(0, 2), (1, 2), (2, 2)],  # Column 3
    [(0, 0), (1, 1), (2, 2)],  # Diagonal 1
    [(0, 2), (1, 1), (2, 0)]   # Diagonal 2
]

# Initialize a black canvas for the board
board_size = 600  # Size of the board
cell_size = board_size // 3  # Size of each cell
black_img = np.zeros((board_size, board_size, 3), dtype=np.uint8)

# Function to draw the board grid
def draw_board():
    global black_img
    # Draw vertical lines
    for i in range(1, 3):
        cv2.line(black_img, (i * cell_size, 0), (i * cell_size, board_size), (255, 255, 255), 5)
    # Draw horizontal lines
    for i in range(1, 3):
        cv2.line(black_img, (0, i * cell_size), (board_size, i * cell_size), (255, 255, 255), 5)

# Function to draw X (two diagonal lines)
def draw_X(x, y):
    # Draw two diagonal lines to form 'X'
    cv2.line(black_img, (x * cell_size + 20, y * cell_size + 20), ((x + 1) * cell_size - 20, (y + 1) * cell_size - 20), (255, 255, 255), 10)
    cv2.line(black_img, ((x + 1) * cell_size - 20, y * cell_size + 20), (x * cell_size + 20, (y + 1) * cell_size - 20), (255, 255, 255), 10)

# Function to draw O (a circle)
def draw_O(x, y):
    # Draw a circle to represent 'O'
    center = (x * cell_size + cell_size // 2, y * cell_size + cell_size // 2)
    cv2.circle(black_img, center, cell_size // 2 - 20, (255, 255, 255), 10)

# Function to check for a winner
def check_winner():
    for combination in winning_combinations:
        values = [board[x][y] for x, y in combination]
        if values[0] != '' and values[0] == values[1] == values[2]:
            return values[0]
    return None

# Function to check for a draw
def check_draw():
    for row in board:
        if '' in row:
            return False
    return True

# Function to display message
def show_message(message):
    global black_img
    # Draw white rectangle for pop-up
    cv2.rectangle(black_img, (board_size // 8, board_size // 3), (board_size * 7 // 8, board_size * 2 // 3), (255, 255, 255), -1)
    cv2.putText(black_img, message, (board_size // 4 - 30, board_size // 2), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 140, 255), 3, cv2.LINE_AA)
    cv2.imshow("Tic-Tac-Toe", black_img)
    cv2.waitKey(0)

# Mouse callback function
def mouse_callback(event, x, y, flags, param):
    global current_player
    if event == cv2.EVENT_LBUTTONDOWN:  # Left click (X)
        row = y // cell_size
        col = x // cell_size

        if board[row][col] == '':  # Only place mark if the cell is empty
            board[row][col] = 'X'
            # Draw the mark
            draw_X(col, row)
            # Check for winner
            winner = check_winner()
            if winner:
                show_message(f"{winner} wins!")
                cv2.waitKey(0)
                return
            # Check for draw
            elif check_draw():
                show_message("It's a draw!")
                cv2.waitKey(0)
                return
            # Switch players
            current_player = 'O'
    
    elif event == cv2.EVENT_RBUTTONDOWN:  # Right click (O)
        row = y // cell_size
        col = x // cell_size

        if board[row][col] == '':  # Only place mark if the cell is empty
            board[row][col] = 'O'
            # Draw the mark
            draw_O(col, row)
            # Check for winner
            winner = check_winner()
            if winner:
                show_message(f"{winner} wins!")
                cv2.waitKey(0)
                return
            # Check for draw
            elif check_draw():
                show_message("It's a draw!")
                cv2.waitKey(0)
                return
            # Switch players
            current_player = 'X'

# Function to reset the game
def reset_game():
    global board, black_img, current_player
    board = [['' for _ in range(3)] for _ in range(3)]
    black_img = np.zeros((board_size, board_size, 3), dtype=np.uint8)
    current_player = 'X'
    draw_board()

# Set up the OpenCV window
cv2.namedWindow('Tic-Tac-Toe')
cv2.setMouseCallback('Tic-Tac-Toe', mouse_callback)

# Game loop
while True:
    # Reset the board state
    black_img = np.zeros((board_size, board_size, 3), dtype=np.uint8)
    draw_board()

    # Redraw the X and O marks on the board
    for row in range(3):
        for col in range(3):
            if board[row][col] == 'X':
                draw_X(col, row)
            elif board[row][col] == 'O':
                draw_O(col, row)

    # Display the current board
    cv2.imshow("Tic-Tac-Toe", black_img)

    # Wait for a key event
    key = cv2.waitKey(1) & 0xFF
    if key == 27:  # ESC key to exit
        break
    elif key == ord('r'):  # 'r' key to reset
        reset_game()

# Cleanup
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import math

# Global variables
spinner_center = (300, 300)  # Center of the spinner
spinner_radius = 100  # Radius of the spinner
spinner_angle = 0  # Current angle of the spinner
spinner_speed = 5  # Initial speed of rotation (degrees per frame)
running = False  # Spinner active status

# Mouse callback function
def mouse_callback(event, x, y, flags, param):
    global running, spinner_speed
    if event == cv2.EVENT_LBUTTONDOWN:  # Left click to start/stop the spinner
        running = not running
    elif event == cv2.EVENT_RBUTTONDOWN:  # Right click to increase spinner speed
        spinner_speed += 5

# Initialize a black canvas
canvas_size = 600
black_img = np.zeros((canvas_size, canvas_size, 3), dtype=np.uint8)

# Set up the OpenCV window
cv2.namedWindow('Loading Spinner')
cv2.setMouseCallback('Loading Spinner', mouse_callback)

while True:
    # Clear the canvas
    black_img[:] = (0, 0, 0)

    # Draw spinner circle outline
    cv2.circle(black_img, spinner_center, spinner_radius, (255, 255, 255), 2)

    # Draw spinner line
    x_end = int(spinner_center[0] + spinner_radius * math.cos(math.radians(spinner_angle)))
    y_end = int(spinner_center[1] - spinner_radius * math.sin(math.radians(spinner_angle)))
    cv2.line(black_img, spinner_center, (x_end, y_end), (0, 255, 0), 4)

    # Update spinner angle if running
    if running:
        spinner_angle = (spinner_angle + spinner_speed) % 360

    # Add text
    status_text = "Running" if running else "Paused"
    cv2.putText(black_img, f"Status: {status_text}", (10, 580), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
    cv2.putText(black_img, "Left Click: Start/Stop | Right Click: Speed Up", (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)

    # Display the spinner
    cv2.imshow("Loading Spinner", black_img)

    # Break on pressing ESC
    key = cv2.waitKey(10)
    if key == 27:  # ESC key
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import time

# Parameters for the spinner
frame_size = 500
center = (frame_size // 2, frame_size // 2)
radius = 100
thickness = 10
font = cv2.FONT_HERSHEY_SIMPLEX

# Frames list to store the animation
frames = []

# Create frames for the animation
for percent in range(15, 56, 10):  # Oscillating percentage values
    frame = np.zeros((frame_size, frame_size, 3), dtype=np.uint8)  # Black frame
    
    # Draw the spinner arc
    start_angle = -90  # Top position
    end_angle = -90 + int(percent * 3.6)  # Map percentage to degrees (360 degrees -> 100%)
    cv2.ellipse(frame, center, (radius, radius), 0, start_angle, end_angle, (0, 255, 0), thickness)
    
    # Draw the static text
    text = f"Loading... {percent}%"
    text_size = cv2.getTextSize(text, font, 1, 2)[0]
    text_x = center[0] - text_size[0] // 2
    text_y = center[1] + text_size[1] + 50
    cv2.putText(frame, text, (text_x, text_y), font, 1, (255, 255, 255), 2, cv2.LINE_AA)
    
    # Add the frame to the list
    frames.append(frame)

# Reverse the frames to loop back
frames += frames[::-1][1:]  # Avoid duplicating the last frame

# Display the animation in a loop
while True:
    for frame in frames:
        cv2.imshow("Loading Spinner", frame)
        if cv2.waitKey(100) & 0xFF == 27:  # Press 'ESC' to exit
            break
    else:
        continue
    break

cv2.destroyAllWindows()
```
